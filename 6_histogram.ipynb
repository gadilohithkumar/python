# Histogram

A histogram is a chart that plots the distribution of a numeric variable’s values as a series of bars. Each bar typically covers a range of numeric values called a bin or class; a bar’s height indicates the frequency of data points with a value within the corresponding bin.

### When you should use a histogram
Histograms are good for showing general distributional features of dataset variables. You can see roughly where the peaks of the distribution are, whether the distribution is skewed or symmetric, and if there are any outliers.

### Difference between histogram and bar plot ?

HISTOGRAM

Histogram refers to a graphical representation, that displays data by way of bars to show the frequency of numerical data.

Distribution of non-discrete variables


BAR GRAPH

Bar graph is a pictorial representation of data that uses bars to compare different categories of data.

Comparison of discrete variables


```python
import matplotlib.pyplot as plt
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]
```

### 1. Plotting the histogram

```python
plt.hist(ages)
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```

### 2. Adding bins to the plot

```python
plt.hist(ages, bins=5)
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```

### 3. Adding edge color

```python
plt.hist(ages, bins=5, edgecolor='black')
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```

### 4. coustomized bins

```python
bins = [10, 20, 30, 40, 50, 60]

plt.hist(ages, bins=bins, edgecolor='black')
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```

### 6. setting log values

```python
plt.hist(age, bins=bins, edgecolor='black',log=True)
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```

### 7. axis vertical line

```python
median_age = 29
color = '#fc4f30'

plt.hist(age, bins=bins, edgecolor='black',log=True)
plt.axvline(median_age, color=color)
plt.xlabel("Ages")
plt.ylabel("Total Respondents")
plt.title("Ages of Respondents")
plt.show()
```
